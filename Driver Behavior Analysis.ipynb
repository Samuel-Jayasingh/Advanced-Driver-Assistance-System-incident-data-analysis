```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'alert_data.csv' with the actual file path to your CSV dataset
csv_file = '/content/unnati_phase1_data_revised.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(csv_file)

# Calculate the correlation between vehicle speed and alerts
correlation = df['Speed'].corr(df['Alert'].astype('category').cat.codes)

# Create a new column 'Speed_Difference' to calculate the difference from the allowed speed (45 mph)
df['Speed_Difference'] = df['Speed'] - 45

# Create a scatterplot to visualize the relationship between speed difference and alert type
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Speed_Difference', y=df['Alert'].astype('category').cat.codes)
plt.title('Driver Behavior Analysis: Speed Difference vs. Alert')
plt.xlabel('Speed Difference from Allowed Speed (mph)')
plt.ylabel('Alert Type')
plt.yticks(ticks=df['Alert'].astype('category').cat.codes.unique(), labels=df['Alert'].unique())
plt.xticks(rotation=0)

# Display the correlation coefficient on the plot
plt.annotate(f'Correlation: {correlation:.2f}', xy=(0.7, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()
```
