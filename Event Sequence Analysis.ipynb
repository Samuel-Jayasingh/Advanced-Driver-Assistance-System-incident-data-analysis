```python
import pandas as pd

# Replace 'your_dataset.csv' with the actual file path to your dataset
dataset_file = '/content/unnati_phase1_data_revised.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_file)

# Combine the 'Date' and 'Time' columns into a single datetime column
df['Date_Time'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Define the sequence you want to analyze (e.g., cas_ldw followed by cas_fcw)
sequence_to_analyze = ['cas_ldw', 'cas_fcw']

# Initialize variables to keep track of the sequence occurrences
sequence_count = 0
total_events = len(df)

# Iterate through the dataset to find the sequence occurrences
for index, row in df.iterrows():
    if row['Alert'] == sequence_to_analyze[sequence_count]:
        sequence_count += 1

        # Check if the entire sequence has been found
        if sequence_count == len(sequence_to_analyze):
            print(f"Sequence found at index {index - len(sequence_to_analyze) + 1} to {index}.")

            # Optional: You can extract and print additional information about the sequence here

            # Reset the sequence counter
            sequence_count = 0
    else:
        # Reset the sequence counter if the sequence is broken
        sequence_count = 0

# If you want to count the total number of sequences found
# total_sequences_found = total_events // len(sequence_to_analyze)
# print(f"Total sequences found: {total_sequences_found}")
```
