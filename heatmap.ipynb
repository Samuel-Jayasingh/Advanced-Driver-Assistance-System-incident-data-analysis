```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Replace 'your_dataset.csv' with the actual file path to your dataset
dataset_file = '/content/unnati_phase1_data_revised.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_file)

# Create a Folium map centered at an initial location (you can adjust this)
m = folium.Map(location=[df['Lat'].mean(), df['Long'].mean()], zoom_start=10)

# Create a HeatMap layer using the geographical coordinates
heat_data = [[row['Lat'], row['Long']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

# Save the map as an image (PNG)
m.save('heatmap.html')
m.save('heatmap.jpg')
```
