```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_dataset.csv' with the actual file path to your dataset
dataset_file = '/content/unnati_phase1_data_revised.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_file)

# Combine the 'Date' and 'Time' columns into a single datetime column
df['Date_Time'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Extract the alert type (cas_fcw, cas_pcw, cas_ldw, cas_hmw) from the 'Alert' column
df['Alert_Type'] = df['Alert'].str.split('_').str[-1]

# Count the frequency of each alert type
alert_counts = df['Alert_Type'].value_counts()

# Plot the frequency of alert types
plt.figure(figsize=(8, 6))
alert_counts.plot(kind='bar', color='skyblue')
plt.title('Alert Frequency Analysis')
plt.xlabel('Alert Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()

# Print the most common alert type
most_common_alert = alert_counts.idxmax()
print(f"The most common alert type is {most_common_alert} with {alert_counts[most_common_alert]} occurrences.")
```
